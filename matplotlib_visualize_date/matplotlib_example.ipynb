```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
import seaborn
```

```python
# matplotlib 한글 표시 설정
import matplotlib.font_manager as fm

# matplotlib의 글꼴 관리자에 특정 글꼴을 추가
fe = fm.FontEntry(
    fname=r'font/NanumGothic.ttf',  # 추가할 글꼴
    name='NanumGothic'              # 글꼴의 이름
)
fm.fontManager.ttflist.insert(0,fe)

# pyplot에 matplotlib의 글꼴 관리자에 등록된 글꼴 적용
plt.rcParams.update({'font.family':'NanumGothic','font.size':12})
# pyplot에서 '-' 를 문자로 취급하도록 설정 _ ? : 데이터 표현의 유연성 , plot(그래프)의 가독성 향상
plt.rcParams['axes.unicode_minus'] = False
```

#### <span style='background-color:rgba(55, 255, 50, 0.3);'>seaborn 을 활용한 1차원 분포 플롯</span>
`seaborn 패키지`에서 제공해주는 여러가지 메서드를 활용하여 간단하게 1차원 분포 플롯을 표시할 수 있음

##### <span style='background-color:rgba(55, 255, 180, 0.5);'>1차원 실수 분포 플롯</span>
`seaborn 패키지`는 히스토그램 형식과 다른 커널 밀도, 러그 플롯을 재공함
- `rugplot()`: 데이터 위치를 x축 위에 작은 선분으로 표시 (러그 플롯)
- `kdeplot()`: 히스토그램과 비슷하지만 곡선형태로 표시(커널 플롯)

```python
data = iris.petal_length.values
data
#실수로 구성되어있는 1차원 배열
```

```python
# 러그 플롯
seaborn.rugplot(data) #1차원 데이터를 rugplot에 전달
plt.title('러그 플롯')
plt.grid(True)
plt.show()
```

```python
# 커널 밀도 플롯
seaborn.kdeplot(data) #1차원 데이터를 kdeplot에 전달
plt.title('커널 밀도 플롯')
plt.grid(True)
plt.show()
```

##### <span style='background-color:rgba(255, 255, 10, 0.3);'>seaborn 을 활용한 다차원 데이터
다차원 데이터는 3가지의 경우의 수를 가질 수 있음
- 모든 데이터가 실수형
- 모든 데이터가 카테고리형
- 데이터가 실수형과 카테고리형이 섞여 있음

##### <span style='background-color:rgba(255, 155, 10, 0.5);'>2차원 실수형 데이터</span>
데이터가 2차원이면서 모두 실수이면 `jointplot()` 메서드를 사용하여 표현 할 수 있음  
`jointplot()` : 스캐터 플롯, 등고선 플롯 등을 나타낼 수 있고 양측에 히스토그램을 표현해 줌 (데이터프레임만 사용 가능)

```python
seaborn.jointplot(x="sepal_length", y="sepal_width", data=iris)
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02)
plt.grid(True)
plt.show()
```

```python
seaborn.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot 과 Kernel Density Plot", y=1.02)
plt.grid(True)
plt.show()
```

##### <span style='background-color:rgba(255, 155, 10, 0.5);'>다차원 실수형 데이터</span>
3차원 이상의 데이터라면 `pairplot()` 메서드로 그리드 형태의 스캐터 플롯의 행렬을 만들어 줌  
그런데, 같은 데이터가 만나는 대각선에는 각 데이터의 히스토그램을 표현

```python
seaborn.pairplot(iris)
plt.suptitle('다차원 Pair plot', y=1.05)
plt.show()
```

카테고리형 데이터가 포함되어 있는 경우라면 `hue`인수에 카테고리 변수 이름을 지정함

```python
seaborn.pairplot(iris, hue='species', markers=['o','s','D'])
#                    iris의 특정 데이터 / 각각의 색 지정
plt.suptitle('다차원 Pair plot', y=1.05)
plt.show()
```

##### <span style='background-color:rgba(255, 155, 10, 0.5);'>2차원 카테고리 데이터</span>
데이터가 2차원이면서 모두 카테고리 값이면 `heatmap()` 메서드로 히트맵을 표현할 수 있음

```python
titanic_size = titanic.pivot_table(
     index="class", columns="sex", aggfunc="size"
)
titanic_size
```

```python
seaborn.heatmap(titanic_size, cmap=seaborn.light_palette("gray", as_cmap=True), annot=True, fmt="d")
#                                                                                           표현 행태
plt.title("히트맵")
plt.show()
```

##### <span style='background-color:rgba(255, 155, 10, 0.5);'>2차원 복합 데이터</span>
데이터가 2차원 이면서 실수 값과 카테고리 값이 섞여있다면 기존 플롯 외의 다른 플롯을 사용할 수 있음  
- `barplot`
- `boxplot`
- `pointplot`
- `violinplot`
- `stripplot`
- `swarmplot`

1) - `barplot()` : 카테고리 값에 따른 실수 값의 평균과 표준 편차를 바 형태로 표시 / 평균은 막대 그래프로, 편차는 에러바로 표현

```python
seaborn.barplot(data=tips, x='day', y='total_bill')
plt.title('바 플롯')
plt.grid(True)
plt.show()
```

2) - `boxplot()` : 사분위수와 아웃라이어라는 데이터를 표현하는 방법

```python
seaborn.boxenplot(data=tips, x='day', y='total_bill')
plt.title('박스 플롯')
plt.grid(True)
plt.show()
```

3) - `violinplot()` : 세로 방향으로 커널 밀도 히스토그램을 그려줌

```python
seaborn.violinplot(data=tips, x='day', y='total_bill')
plt.title('바이올릿 플롯')
plt.grid(True)
plt.show()
```

4) - `stripplot()` : 모든 데이터를 점으로 표시해서 보여줌

```python
seaborn.stripplot(data=tips, x='day', y='total_bill')
plt.title('스트립 플롯')
plt.grid(True)
plt.show()
```

5) - `swarmplot()` : 스트립 플롯에서 데이터가 겹치는 부분을 겹치지 않게 표현

```python
seaborn.swarmplot(data=tips, x='day', y='total_bill')
plt.title('스웜 플롯')
plt.grid(True)
plt.show()
```

##### <span style='background-color:rgba(255, 55, 150, 0.5);'>seaborn을 활용한 스타일 변경</span>
seaborn 패키지는 matplotlib 에서 제공하는 기본 스타일이 아닌 특정 스타일로 변경할 수 있도록 함  
- `set()`메서드로 색상, 틱 스타일 등을 seaborn의 기본 스타일로 변경  
- `set_style()`메서드로 틱 스타일 변경
1. - darkgrid 
2. - whitegrid 
3. - dark 
4. - white
5. - ticks

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for number in range(1,7):
        plt.plot(x, np.sin(x + number * 0.5)  * (7 - number * flip))
    plt.show()
```

```python
seaborn.set_style('ticks') #ticks : 숫자 옆에 - 표시
sinplot()
```

```python
seaborn.set_style('darkgrid') #배경
sinplot()
```
