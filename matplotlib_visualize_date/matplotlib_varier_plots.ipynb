##### <span style='background-color:rgba(155, 155, 150, 0.5);'>바 차트 (막대 그래프)</span>
x축의 값이 분류가 가능한 카테고리 형태일 때 `bar()` , `barh()` 메서드로 막대그래프를 표시할 수 있음  
주의! bar 메서드는 x축의 값이 첫 번째 인수로 옴 

````python
y = [2,3,1]
x = [1,2,3]
xlabel = ['가', '나', '다']

plt.bar(x,y)
plt.xticks(x,xlabel)
plt.yticks(sorted(y))
plt.show()
````

`xerr` , `yerr` 인수를 지정하면 에러 바를 추가할 수 있음  
`alpha` 인수를 지정하면 바의 투명도를 지정할 수 있음 (0:투명 / 1:불투명)  

````python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_position = np.arange(4)
y_performance = (3+10) * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar h')
plt.barh('y_position' , y_performance , xerr = error, alpha = 0.4)
plt.yticks(y_position, people)
plt.xlabel('x축')
plt.grid(True)
plt.show()
````

만약 x에 대하여 여러 변수를 한 차트로 비교하고자 할 때 점의 크기 혹은 색상으로 추가 표현할 수 있음

##### <span style='background-color:rgba(155, 155, 150, 0.5);'>컨투어 플롯</span>
3차원 자료에서 x,y 입력과 z출력이 존재할 때, z 표현을 명암이 아닌 선을 사용하는 차트  
`contour` , `contourf` 메서드를 사용  
`contour` : 등고선만 표시  
`contourf` : 색으로 표시  
입력 x y 에 대하며 1차원 배열을 그대로 사용할 수 없고, `meshgrid()` 메서드로 포인트 행렬을 만들어 사용해야 함

````python
def f(x,y):
    return (1 -x /2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
xx, yy = np.meshgrid(x, y)
zz = f(xx, yy)

plt.title('컨투어 플롯')
plt.contour(xx,yy,zz, colors='blue')   #?***
plt.grid(True)
plt.show()
````

##### <span style='background-color:rgba(155, 155, 150, 0.5);'>3D 서피스 플롯</span>
3차원 데이터를 입체적으로 표현하는 차트  
3d Axes 객체를 생성하여 사용함  
`plot_wireframe()` , `plot_surface()` 메서드로 표현

````python
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
xx, yy = np.meshgrid(x, y)
rr = np.sqrt(xx**2 + yy**2) #빗변 구하는 식
zz = np.sin(rr) #sin 함수 이용
  
axes = plt.axes(projection="3d")
axes.set_title('3D 서피스 플롯')
axes.plot_surface(xx, yy, zz, rstride=1, cstride=1, cmap='hot')                         
plt.show()
````

##### <span style='background-color:rgba(155, 55, 50, 0.5);'>삼각 그리드 생성</span>
`Triangulation` 클래스로 삼각 그리드를 생성할 수 있음  
`Triangulation` 클래스의 생성자는 x,y,triangles 세 개의 인수를 받음  
`triangles`는 각 점의 기하학적 위상 정보

````python
x = np.array([0, 1, 2])
y = np.array([0, np.sqrt(3), 0])
triangles = [[0, 1, 2]] #2차원
triangle = mtri.Triangulation(x, y, triangles)

plt.title("삼각 그리드")
plt.triplot(triangle, 'ko-')
plt.xlim(-0.1, 2.1)
plt.ylim(-0.1, 1.8)
plt.grid(True)
plt.show()
````

````python
# 여러개의 삼각형 그리기
x = np.asarray([0, 1, 2, 3, 4, 2])
y = np.asarray([0, np.sqrt(3), 0, np.sqrt(3), 0, 2*np.sqrt(3)])
triangles = [[0, 1, 2], [2, 3, 4], [1, 2, 3], [1, 3, 5]]
triangle = mtri.Triangulation(x, y, triangles)

plt.title("여러개의 삼각형 그리기")
plt.triplot(triangle, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.grid(True)
plt.show()
````

##### <span style='background-color:rgba(155, 55, 50, 0.5);'>삼각 그리드 세분화</span>
`TriRefiner` , `UniformTriRefiner` 클래스로 그리드를 세분화 할 수 있음  
- `refine_triangulation()` : 단순히 삼각 그리드를 세분화
- `refine_field()` : 실제 함수에 최적화된 삼각 그리드 생성

````python
refiner = mtri.UniformTriRefiner(triangle)
triangle2 = refiner.refine_triangulation(subdiv=1) #subdiv : 하위구분 / 자르는 수?

plt.title("삼각 그리드 세분화")
plt.triplot(triangle2, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.grid(True)
plt.show()
````

##### <span style='background-color:rgba(155, 55, 50, 0.5);'>삼각 그리드 플롯</span>
삼차원 그리드 상에 `tricontour` , `tricontourf` , `plot_trisur` , `plot_trisurf` 를 사용하여 2차원 등고선이나 3차원 표면을 그릴 수 있음

````python
triangle3 = refiner.refine_triangulation(subdiv=5)
z = np.cos(1.5 * triangle3.x) * np.cos(1.5 * triangle3.y)
plt.title("삼각 그리드 등고선")
plt.tricontourf(triangle3, z, cmap="jet")
plt.grid(True)
plt.show()
````

````python
triangle4 = refiner.refine_triangulation(subdiv=3)
z = np.cos(1.5 * triangle4.x) * np.cos(1.5 * triangle4.y)

axes = plt.axes(projection='3d')
axes.set_title('삼각 그리드 3차원')
axes.plot_trisurf(triangle4.x, triangle4.y, z, cmap='jet', linewidth=0.2)
axes.set_zlim(-1, 1)
axes.view_init(40, -40)
plt.show()
````
