##### <span style='background-color:rgba(215, 15, 150, 0.5);'>차트의 종류</span>

_**<u>막대 그래프</u>**_ : 데이터의 카테고리별 값을 막대로 나타내는 방식(수직 막대 그래프, 수평 막대 그래프)  
> 카테고리별로 비교할 때 주로 사용됨 (제품별 판매량, 지역별 인구수 ..)
    
_**<u>선 그래프</u>**_ : 시간의 흐름에 따른 데이터 포인터를 선으로 연결하여 나타내는 방식  
> 시간의 흐름에 따라 비교할 때 주로 사용됨 (기온의 변화, 주식 가격 ..) _ 단순히 값 자체를 강조 (<->영역 그래프와 비교)
  
_**<u>파이 차트</u>**_ : 전체에 대한 비율을 원형으로 나타내는 방식 (각 조각의 크기 = 해당 부분의 비율)  
> 전체에 대한 비율을 비교할 때 주로 사용됨 (여론조사 결과 ..)  
  
_**<u>산점도</u>**_ : 두 변수 간의 상관 관계를 점들의 분포로 나타내는 방식
> 두 변수의 상관 관계를 파악할 때 주로 사용 (키와 몸무게의 상관관계 ..)

_**<u>히스토그램</u>**_ : 범위별로 그룹화하여 각 그룹애 속하는 데이터 포인터의 수를 막대로 나타내는 방식
> 데이터의 분포를 비교할 때 주로 사용됨 (시험 점수의 분포 ..)

_**<u>영역 그래프</u>**_ : 선 그래프의 아래 영역에 색으로 채운 형태
> 시간의 흐름에 따라 비교할 때 주로 사용됨, 변화량을 강조할 때 주로 사용됨 (누적 데이터 ..)

_**<u>박스 그래프</u>**_ : 데이터의 분포를 나타내는 방식, 최솟값, 최댓값, 사분위를 나타내는 방식
> 데이터의 이상치를 판단할 때 주로 사용됨 (주식 ..)

##### <span style='background-color:rgba(5, 155, 150, 0.5);'>차트 그림의 구조</span>
matplotlib 으로 그려지는 차트는 Figure 객체, Axes 객체, Axis 객체로 구성이 되어져 있음  
Figure 객체는 하나 이상의 Axes 객체를 포함하고 있고,  
Axes 객체는 두 개 이상의 Axis 객체를 포함하고 있음  
- `Figure` : 차트가 그려지는 캔버스
- `Axes` : 하나의 플룻(그래프)
- `Asix` : 하나의 축

##### <span style='background-color:rgba(25, 155, 50, 0.5);'>Figure 객체</span>
- 모든 차트 그래프 결과는 Figure 객체에 포함되어 있음  
- 내부 차트가 아닌 경우 하나의 Figure 객체에는 하나의 아이디와 원도우를 가짐  
- 일반적으로 기본 파이썬에서 차트를 표시하면 Figure 하나당 하나의 원도우를 가짐  
- `plot()` 메서드가 자동으로 Figure 객체를 생성하기 때문에 직접 생성하는 경우는 드물지만   
여러 개의 원도우를 동시에 출력할 때나 차트 전체의 크기를 변경하고 싶을 때 사용할 수 있음 

````python
# Figure 만들기
plt.figure(figsize=(10,2))
plt.title('figure 객체 사이즈 변경')
plt.plot(np.random.randn(100))
plt.show()
````

현재 사용하고 있는 Figure 객체를 얻으려면 `gcf()` 메서드로 얻을 수 있음

````python
f1 = plt.figure(figsize=(10,2))
plt.title('figure 객체 사이즈 변경')
plt.plot(np.random.randn(100))

f2 = plt.gcf()
print(id(f1))
print(id(f2))     # 같은 아이디를 가진 것을 알 수 있다
plt.show()

plt.title('figure 객체 사이즈 변경')
plt.plot(np.random.randn(100))

f3 = plt.gcf()
print(id(f3))  # 새로운 Figure 생성 / 다른 아이디를 가진다.
````

##### <span style='background-color:rgba(25, 155, 50, 0.5);'>`Axes` 객체와 subplot</span>
하나의 Figure 객체 안에 여러개의 Axes 객체를 생성하여 행렬 형태의 서브 플롯을 생성할 수 있음  
이떄, `subplot()` 메서드를 사용하여 `Axes` 명시적으로 객체를 사용할 수 있음  
그러나, `pyplot`의 `plot()`메서드를 바로 사용해도 `Axes` 객체를 생성할 수 있음

- +)  이 때 `subplot()` 의 인수는 (2,2,1)를 줄여서 (221) 라는 하나의 숫자로 표시할 수도 있다  
Axes의 위치는 위에서 부터 아래로, 왼쪽에서 오른쪽으로 카운트한다. 

````python
x1 = np.linspace(0,5)
x2 = np.linspace(0,2)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1)
plt.plot(x1, y1, 'yo-')
plt.title('2행 1열 서브 플롯 첫번째')

plt.subplot(2,1,2)
plt.plot(x2, y2, 'r.-')
plt.title('2행 1열 서브 플롯 두번째')

plt.tight_layout()  # 겹치는 부분의 간격을 자동으로 맞춰줌
plt.show()
````

`subplot()` 메서드의 (x, y, index) 의 인수를 (xyindex)로 붙여서 사용할 수 있음  
ex) `subplot(2, 2, 1)` -> `subplot(221)` 

````python
plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title('첫번째 플롯')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('두번째 플롯')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('세번째 플롯')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('네번째 플롯')

plt.tight_layout()
plt.show()
````

`subplot(x, y)` 메서드로 복수의 Axes 객체를 담은 배열을 생성할 수 있음  
`subplot(x, y)` 메서드는 (figure, axes 배열) 튜플을 반환

````python
figure, axes = plt.subplots(2, 2)

axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title('1행 1열')

axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title('1행 2열')

axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title('2행 1열')

axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title('2행 2열')

plt.tight_layout()
plt.show()
````

##### <span style='background-color:rgba(25, 155, 50, 0.5);'>`Axes` 객체와 축</span>
`twinx()`메서드로 x축을 공유하는 새로운 Axes객체를 생성할 수 있음

````python
figure, axes0 = plt.subplots()
axes1 = axes0.twinx() # x축이 같은 새로운 axes 생성

axes0.set_title('x축을 공유하는 플롯')
axes0.plot([10,5,2,9,7], 'r-', label='y0')
axes0.set_ylabel('y0')

axes1.plot([100,200,220,180,120], 'g:', label='y1')
axes1.set_ylabel('y1')

axes0.set_xlabel('x축')
plt.show()
````
